```python
# latihan 

# dari data nba tadi
# gunakan query untuk mencari game milik tim Boston Celtics saja (team id BOS, team Celtics)
# hitung rata2 pts yang di skor tim ini

```

```python
df_nba_small
```

```python
kondisi1 = (df_nba_small["team_id"] == "BOS")
kondisi2 = (df_nba_small["team"] == "Celtics")

df_bos = df_nba_small[kondisi1 & kondisi2]
df_bos
```

```python
df_bos["team_pts"].mean()
```

```python
# visualisasi data pts yang dicetak tim BOS dari tahun ke tahun

# bikin copy dataframe
df_bos2 = df_bos.copy()

# bikin kolom tahun
df_bos2["year"] = df_bos["date"].dt.year
df_bos2
```

```python
# grup data dari tahun ke tahun

df_bos_yearly = df_bos2.groupby("year")[["team_pts"]].sum()
df_bos_yearly
```

```python
# bikin visualisasi
df_bos_yearly.plot()
```

```python

df_bos_yearly.plot(marker="o", color="green", figsize=(12,6))
```

```python
# latihan

# coba cari untuk tim LAL Lakers
# total skor pts dari tahun ke tahun
# buat grafik seperti di atas
# kasih warna merah

kondisi = (df_nba_small["team_id"] == "LAL")

df_lal = df_nba_small[kondisi]

df_lal["tahun"] = df_nba_small["date"].dt.year

df_lal_yearly = df_lal.groupby(["tahun"])[['team_pts']].sum()

df_lal_yearly.plot()
```

```python
df_lal_yearly.info()
```

```python
df_lal_yearly.plot(marker="o", color="purple", figsize=(12,6))
```

```python
print(type(df_bos_yearly), type(df_lal_yearly))
print(df_bos_yearly.shape, df_lal_yearly.shape)
```

```python
# gabung dua dataframe jadi satu

df_bos_yearly.columns = ["BOS"]
df_lal_yearly.columns = ["LAL"]

df_bos_lal = pd.concat([df_bos_yearly, df_lal_yearly], axis=1)
df_bos_lal
```
