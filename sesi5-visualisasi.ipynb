```python
# kita akan import data imigrasi seluruh dunia ke Canada per tahun
# menggunakan pandas
df_can = pd.read_excel("dataset/Canada.xlsx",
            sheet_name="Canada by Citizenship",
            skiprows=20,
            skipfooter=2)

df_can
```

```python
df_can.info()
```

```python
df_can.columns
```

```python
# drop kolom tidak terpakai
 
df_can_dropped = df_can.drop(columns=[ 'Type', 'Coverage', 'AREA', 'AreaName',      'REG',
        'RegName',      'DEV',  'DevName'])
```

```python
df_can_dropped.info()
```

```python
# ubah nama kolom OdName jadi Country
df_can_dropped = df_can_dropped.rename(columns={"OdName":"Country"})

df_can_dropped.head()
```

```python
# jadikan index kolom nama negaranya
df_can_dropped = df_can_dropped.set_index("Country")
```

```python
df_can_dropped.info()
```

```python
# line plot
# plot yang simple, garis
# biasanya kita tambahkan opsi warna color, marker
# biasanya tidak lebih dari 5, paling banyak 7 garis di satu plot

df_indosg = df_can_dropped.loc[["Indonesia", "Singapore"]]
df_indosg.plot()
```

```python
df_indosg.T.plot()
```

```python
df_indosg.T.plot(figsize=(12,6), marker="o")

plt.title("Imigrasi Indonesia & Singapore ke Canada 1980-2013")
plt.xlabel("tahun")
plt.ylabel("total imigran")
```

```python
# informasi yang didapat jangan lupa ditulis
# untuk Singapore, jumlah imigran yang pergi ke Canada naik drastis di tahun 1987-1989
# sementara untuk Indonesia, kenaikan jumlah imigran terjadi di tahun 1999-2000
```
